```python
# Load Qiskit
from qiskit_ibm_runtime import QiskitRuntimeService, Batch
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv()
API_TOKEN = os.getenv("API_TOKEN")
API_INSTANCE = os.getenv("API_INSTANCE", None)
service = QiskitRuntimeService(channel="ibm_quantum", token=API_TOKEN, instance=API_INSTANCE)
backend = service.backend(name="ibm_rensselaer")
```

```python
from qiskit import QuantumCircuit, transpile

qc = QuantumCircuit(5)
qc.initialize('0'*5)
qc.mcx([0, 3], 4)

qc.decompose(reps=1).draw('mpl')
```

```python
rpi_qc = transpile(circuits=qc, backend=backend, optimization_level=3)
rpi_qc.draw('mpl')
```

```python
qc = QuantumCircuit(5)
qc.initialize('0'*5)
qc.cx(0, 1)
qc.cx(3, 2)

qc.mcx([1, 2], 4)

qc.cx(3, 2)
qc.cx(0, 1)


qc.draw('mpl')
```

```python
rpi_qc = transpile(circuits=qc, backend=backend, optimization_level=3)
rpi_qc.draw('mpl')
```
