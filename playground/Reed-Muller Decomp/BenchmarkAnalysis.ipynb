```python
# %% [markdown]
# # Enhanced Scaling Analysis - Exponential Detection + Graph Structure
# Focus on key metrics plus graph structure analysis for exponential scaling patterns

# %% Imports and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
from pathlib import Path
import json
from collections import defaultdict

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')

# %% Data Loading
def load_scaling_data(db_path: str) -> pd.DataFrame:
    """Load the key metrics plus graph structure data for scaling analysis."""
    conn = sqlite3.connect(db_path)
    
    query = """
    SELECT 
        problem_id,
        multiplicative_complexity,
        additive_complexity, 
        longest_path_len,
        max_width_level,
        max_width,
        and_dist_by_level,
        xor_dist_by_level,
        and_count,
        xor_count,
        num_nodes,
        trial_params
    FROM compilation_results
    WHERE success = 1
    """
    
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    # Extract clique size
    df['clique_size'] = df['trial_params'].apply(
        lambda x: json.loads(x).get('clique_size', None)
    )
    
    # Load vertex counts
    problems_db_path = '../../benchmarklib/shared/problems/clique.db'
    if not Path(problems_db_path).exists():
        problems_db_path = 'shared/problems/clique_problems.db'
    
    conn = sqlite3.connect(problems_db_path)
    problems_query = """
    SELECT 
        instance_id as problem_id,
        size_metrics
    FROM problem_instances
    """
    problems_df = pd.read_sql_query(problems_query, conn)
    conn.close()
    
    problems_df['num_vertices'] = problems_df['size_metrics'].apply(
        lambda x: json.loads(x).get('num_vertices', None)
    )
    
    df = df.merge(problems_df[['problem_id', 'num_vertices']], on='problem_id', how='left')
    
    return df

# %% Simple Statistics
def compute_basic_stats(df: pd.DataFrame, metric: str, group_by: str = 'num_vertices'):
    """Compute mean, median, and spread statistics."""
    stats = df.groupby(group_by)[metric].agg([
        'count', 'mean', 'median', 'std', 'min', 'max',
        lambda x: x.quantile(0.25),  # Q1
        lambda x: x.quantile(0.75),  # Q3
    ]).round(2)
    
    stats.columns = ['count', 'mean', 'median', 'std', 'min', 'max', 'q25', 'q75']
    stats['iqr'] = stats['q75'] - stats['q25']
    
    return stats

# %% Gate Distribution Analysis
def compute_gate_level_distributions(df: pd.DataFrame):
    """Compute average gate distributions by level for each vertex count."""
    
    results = {}
    
    for n_vertices in sorted(df['num_vertices'].unique()):
        subset = df[df['num_vertices'] == n_vertices]
        
        # Master histograms for this vertex count
        master_and_hist = defaultdict(int)
        master_xor_hist = defaultdict(int)
        total_graphs = 0
        
        for idx, row in subset.iterrows():
            if pd.notna(row['and_dist_by_level']) and pd.notna(row['xor_dist_by_level']):
                try:
                    and_dist = json.loads(row['and_dist_by_level'])
                    xor_dist = json.loads(row['xor_dist_by_level'])
                    
                    # Add to master histogram
                    for level, count in enumerate(and_dist):
                        master_and_hist[level] += count
                    
                    for level, count in enumerate(xor_dist):
                        master_xor_hist[level] += count
                    
                    total_graphs += 1
                except:
                    continue
        
        # Convert to average distributions
        if total_graphs > 0:
            avg_and_dist = {level: count/total_graphs for level, count in master_and_hist.items()}
            avg_xor_dist = {level: count/total_graphs for level, count in master_xor_hist.items()}
            
            results[n_vertices] = {
                'and_dist': avg_and_dist,
                'xor_dist': avg_xor_dist,
                'total_graphs': total_graphs,
                'max_level': max(max(master_and_hist.keys(), default=0), max(master_xor_hist.keys(), default=0))
            }
    
    return results

# %% Enhanced Plotting Functions
def plot_metric_scaling(df: pd.DataFrame, metric: str, metric_name: str, save_path=None):
    """Create simple linear and log scale plots for a single metric."""
    
    # Compute statistics
    stats = compute_basic_stats(df, metric)
    
    # Create plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(f'{metric_name} vs Number of Vertices', fontsize=14, y=1.02)
    
    x = stats.index
    y_mean = stats['mean']
    y_median = stats['median']
    yerr_std = stats['std']
    yerr_iqr = [stats['median'] - stats['q25'], stats['q75'] - stats['median']]
    
    # Linear scale plot
    ax1.errorbar(x, y_mean, yerr=yerr_std, 
                marker='o', label='Mean  Std', linewidth=2, markersize=6, capsize=4)
    ax1.errorbar(x, y_median, yerr=yerr_iqr,
                marker='s', label='Median (Q1-Q3)', linewidth=2, markersize=6, capsize=4)
    
    ax1.set_xlabel('Number of Vertices')
    ax1.set_ylabel(metric_name)
    ax1.set_title('Linear Scale')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # Log scale plot  
    ax2.errorbar(x, y_mean, yerr=yerr_std,
                marker='o', label='Mean  Std', linewidth=2, markersize=6, capsize=4)
    ax2.errorbar(x, y_median, yerr=yerr_iqr,
                marker='s', label='Median (Q1-Q3)', linewidth=2, markersize=6, capsize=4)
    
    ax2.set_xlabel('Number of Vertices')
    ax2.set_ylabel(f'{metric_name} (log scale)')
    ax2.set_title('Log Scale')
    ax2.set_yscale('log')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    
    return stats

def plot_gate_level_distributions(gate_distributions, save_path=None):
    """Create histograms showing average gate distributions by level."""
    
    vertex_counts = sorted(gate_distributions.keys())
    n_vertices = len(vertex_counts)
    
    fig, axes = plt.subplots(2, n_vertices, figsize=(4*n_vertices, 8))
    if n_vertices == 1:
        axes = axes.reshape(2, 1)
    
    fig.suptitle('Average Gate Distribution by Level', fontsize=16, y=1.02)
    
    # Plot AND distributions
    for i, n_v in enumerate(vertex_counts):
        data = gate_distributions[n_v]
        and_dist = data['and_dist']
        
        levels = sorted(and_dist.keys())
        counts = [and_dist[level] for level in levels]
        
        axes[0, i].bar(levels, counts, alpha=0.7, color='red')
        axes[0, i].set_title(f'{n_v} vertices\n(n={data["total_graphs"]} graphs)')
        axes[0, i].set_xlabel('Level')
        axes[0, i].set_ylabel('Average AND Gates')
        axes[0, i].grid(True, alpha=0.3)
    
    # Plot XOR distributions  
    for i, n_v in enumerate(vertex_counts):
        data = gate_distributions[n_v]
        xor_dist = data['xor_dist']
        
        levels = sorted(xor_dist.keys())
        counts = [xor_dist[level] for level in levels]
        
        axes[1, i].bar(levels, counts, alpha=0.7, color='blue')
        axes[1, i].set_xlabel('Level')
        axes[1, i].set_ylabel('Average XOR Gates')
        axes[1, i].grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

def plot_width_analysis(df: pd.DataFrame, save_path=None):
    """Create scatter plot of max_width_level vs max_width, colored by vertex count."""
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    # Get unique vertex counts for coloring
    vertex_counts = sorted(df['num_vertices'].unique())
    colors = plt.cm.viridis(np.linspace(0, 1, len(vertex_counts)))
    color_map = dict(zip(vertex_counts, colors))
    
    # Create scatter plot
    for n_v in vertex_counts:
        subset = df[df['num_vertices'] == n_v]
        if len(subset) > 0:
            ax.scatter(subset['max_width_level'], subset['max_width'], 
                      c=[color_map[n_v]], label=f'{n_v} vertices', 
                      alpha=0.7, s=50)
    
    ax.set_xlabel('Level of Maximum Width')
    ax.set_ylabel('Maximum Width')
    ax.set_title('Maximum Width vs Level (colored by input size)')
    ax.grid(True, alpha=0.3)
    ax.legend(title='Input Size', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

# %% Print Statistics Summary
def print_metric_summary(stats: pd.DataFrame, metric_name: str):
    """Print basic statistics summary."""
    print(f"\n{metric_name} Summary:")
    print("=" * (len(metric_name) + 9))
    print(f"{'Vertices':<8} {'Count':<6} {'Mean':<12} {'Median':<12} {'Std':<12} {'Min':<12} {'Max':<12}")
    print("-" * 80)
    
    for vertices, row in stats.iterrows():
        print(f"{vertices:<8} {row['count']:<6} {row['mean']:<12.2e} {row['median']:<12.2e} "
              f"{row['std']:<12.2e} {row['min']:<12.2e} {row['max']:<12.2e}")
    
    # Check for exponential-like growth
    if len(stats) > 2:
        mean_ratios = []
        median_ratios = []
        
        for i in range(1, len(stats)):
            if stats.iloc[i-1]['mean'] > 0:
                mean_ratios.append(stats.iloc[i]['mean'] / stats.iloc[i-1]['mean'])
            if stats.iloc[i-1]['median'] > 0:
                median_ratios.append(stats.iloc[i]['median'] / stats.iloc[i-1]['median'])
        
        if mean_ratios:
            avg_mean_ratio = np.mean(mean_ratios)
            avg_median_ratio = np.mean(median_ratios)
            print(f"\nGrowth Analysis:")
            print(f"  Average ratio (mean): {avg_mean_ratio:.2f}")
            print(f"  Average ratio (median): {avg_median_ratio:.2f}")
            
            if avg_mean_ratio > 2:
                print(f"   Potential exponential growth detected (ratio > 2)")
            elif avg_mean_ratio > 1.5:
                print(f"   Fast growth detected (ratio > 1.5)")
            else:
                print(f"   Moderate growth (ratio  1.5)")

def print_gate_distribution_summary(gate_distributions):
    """Print summary of gate distributions."""
    print(f"\n" + "="*60)
    print("GATE LEVEL DISTRIBUTION ANALYSIS")
    print("="*60)
    
    for n_vertices in sorted(gate_distributions.keys()):
        data = gate_distributions[n_vertices]
        print(f"\n{n_vertices} vertices ({data['total_graphs']} graphs):")
        print("-" * 40)
        
        and_dist = data['and_dist']
        xor_dist = data['xor_dist']
        max_level = data['max_level']
        
        print(f"{'Level':<6} {'Avg ANDs':<10} {'Avg XORs':<10}")
        print("-" * 26)
        
        for level in range(max_level + 1):
            and_count = and_dist.get(level, 0)
            xor_count = xor_dist.get(level, 0)
            print(f"{level:<6} {and_count:<10.2f} {xor_count:<10.2f}")
        
        # Total gates
        total_and = sum(and_dist.values())
        total_xor = sum(xor_dist.values())
        print(f"\nTotal avg AND gates: {total_and:.2f}")
        print(f"Total avg XOR gates: {total_xor:.2f}")

# %% Main Analysis Function
def run_enhanced_scaling_analysis(db_path: str, output_dir=None):
    """Run enhanced scaling analysis with graph structure insights."""
    
    print(f"Loading data from: {db_path}")
    df = load_scaling_data(db_path)
    
    print(f"\nLoaded {len(df)} successful compilations")
    print(f"Vertex range: {df['num_vertices'].min()} to {df['num_vertices'].max()}")
    
    if len(df) == 0:
        print("No data found!")
        return df
    
    # Create output directory
    if output_dir:
        Path(output_dir).mkdir(parents=True, exist_ok=True)
    
    # Define the 6 key metrics of interest
    metrics = [
        ('multiplicative_complexity', 'Expression AND Gates'),
        ('additive_complexity', 'Expression XOR Gates'),
        ('longest_path_len', 'Graph Longest Path'),
        ('and_count', 'Graph AND Gates'),
        ('xor_count', 'Graph XOR Gates'),
        ('num_nodes', 'Graph Node Count')
    ]
    
    print("\n" + "="*80)
    print("ENHANCED EXPONENTIAL GROWTH DETECTION")
    print("="*80)
    
    # Analyze each metric
    all_stats = {}
    for metric, metric_name in metrics:
        print(f"\nAnalyzing {metric_name}...")
        
        # Generate plot and get statistics
        save_path = f"{output_dir}/{metric}_scaling.png" if output_dir else None
        stats = plot_metric_scaling(df, metric, metric_name, save_path)
        
        # Print summary
        print_metric_summary(stats, metric_name)
        
        # Store for comparison
        all_stats[metric] = stats
    
    # Gate distribution analysis
    print(f"\n" + "="*80)
    print("COMPUTING GATE LEVEL DISTRIBUTIONS...")
    print("="*80)
    
    gate_distributions = compute_gate_level_distributions(df)
    print_gate_distribution_summary(gate_distributions)
    
    # Plot gate distributions
    gate_save_path = f"{output_dir}/gate_level_distributions.png" if output_dir else None
    plot_gate_level_distributions(gate_distributions, gate_save_path)
    
    # Width analysis
    print(f"\n" + "="*80)
    print("GRAPH WIDTH ANALYSIS...")
    print("="*80)
    
    width_save_path = f"{output_dir}/width_analysis.png" if output_dir else None
    plot_width_analysis(df, width_save_path)
    
    # Print width statistics
    width_stats = compute_basic_stats(df, 'max_width')
    print_metric_summary(width_stats, 'Maximum Width')
    
    # Overall summary
    print(f"\n" + "="*80)
    print("OVERALL SUMMARY")
    print("="*80)
    
    for metric, metric_name in metrics:
        stats = all_stats[metric]
        if len(stats) > 2:
            # Simple growth rate analysis
            first_mean = stats.iloc[0]['mean']
            last_mean = stats.iloc[-1]['mean']
            first_vertices = stats.index[0]
            last_vertices = stats.index[-1]
            
            if first_mean > 0:
                total_growth = last_mean / first_mean
                vertices_span = last_vertices - first_vertices
                avg_growth_per_vertex = total_growth ** (1/vertices_span) if vertices_span > 0 else 1
                
                print(f"\n{metric_name}:")
                print(f"  Total growth factor: {total_growth:.2f}x")
                print(f"  Average growth per vertex: {avg_growth_per_vertex:.2f}x")
                
                if avg_growth_per_vertex > 2:
                    print(f"   EXPONENTIAL GROWTH DETECTED")
                elif avg_growth_per_vertex > 1.5:
                    print(f"   FAST GROWTH")
                else:
                    print(f"   MODERATE GROWTH")
    
    return df, all_stats, gate_distributions

# %% Usage
if __name__ == "__main__":
    # Example usage
    db_path = "project_benchmarks/results-graph-stats.db"
    df, stats, gate_dists = run_enhanced_scaling_analysis(db_path, output_dir="scaling_enhanced")
```
